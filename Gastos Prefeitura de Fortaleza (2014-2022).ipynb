#1. Gastos de 2014 à 2022

##1.1. Gastos por ano (em bilhões)

```python
import pandas as pd
import matplotlib.pyplot as plt
arq = r'https://drive.google.com/u/0/uc?id=18P0ib7LEavG9GOCfBHBAPAGfDHrtjH43&export=download'

gastos = pd.read_excel(arq)

#Exibição dos dados em tabela
gastos['VALOR PAGO'] = gastos['VALOR PAGO'].div(1000000000).round(2)
gastos = gastos[['ANO', 'VALOR PAGO']].groupby(by = 'ANO').agg('sum').reset_index()
gastos.loc['TOTAL'] = gastos.sum()
gastos.loc[gastos.index[-1], 'ANO'] = ''
display(gastos)

gastos = gastos.drop(gastos.index[-1])
plt.subplots(figsize =(8, 5))
plt.margins(0.1)
plt.plot(gastos["ANO"], gastos["VALOR PAGO"], '-bo')
plt.title("Evolução anual de gastos (em bilhões)")

for x,y in zip(gastos["ANO"], gastos["VALOR PAGO"]):
    label = "{:.2f}".format(y)
    plt.annotate(label,(x,y),textcoords="offset points",xytext=(-5,10),ha='center')
plt.show()
```

##1.2. Gastos por área (em bilhões)

```python
import pandas as pd
import matplotlib.pyplot as plt
arq = r'https://drive.google.com/u/0/uc?id=18P0ib7LEavG9GOCfBHBAPAGfDHrtjH43&export=download'

gastos = pd.read_excel(arq)

#Exibição dos dados em tabela
gastos['VALOR PAGO'] = gastos['VALOR PAGO'].div(1000000000).round(2)
gastos = gastos[['CATEGORIA', 'VALOR PAGO']].groupby(by = 'CATEGORIA').agg('sum').sort_values(['VALOR PAGO'], ascending=False).reset_index()
gastos = gastos[gastos['VALOR PAGO'] != 0]
gastos.loc['TOTAL'] = gastos.sum()
gastos.loc[gastos.index[-1], 'CATEGORIA'] = ''
display(gastos)

gastos = gastos.drop(gastos.index[-1])
demais = pd.Series({'CATEGORIA': 'DEMAIS', 'VALOR PAGO': gastos["VALOR PAGO"].sum() - gastos.head(5)["VALOR PAGO"].sum()})
gastos = gastos.head(5)
gastos.loc[len(gastos)] = demais

plt.subplots(figsize =(8, 5))
plt.pie(gastos["VALOR PAGO"],autopct='%1.1f%%', labels = gastos["CATEGORIA"])
plt.title("Gastos por área entre 2014 à 2022")
plt.show()
```
